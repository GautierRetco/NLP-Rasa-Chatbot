```python
import pandas as pd
import ast
# Supposons que df soit votre DataFrame Pandas avec une colonne 'genres' contenant des listes de genres

# Créez un ensemble pour stocker les genres uniques
genres_uniques = set()

# Parcourez chaque ligne du DataFrame pour extraire les genres uniques
for genres_list in df['genres']:
    list = ast.literal_eval(genres_list)

    for genre in list:
        print(genre)
        genres_uniques.add(genre)

print(genres_uniques)
```
